```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.

df_2020 = pd.read_csv("Resources/athletic_sales_2020.csv")
df_2021 = pd.read_csv("Resources/athletic_sales_2021.csv")
```

```python
# Display the 2020 sales DataFrame
df_2020
```

```python
# Display the 2021 sales DataFrame
df_2021
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
df_2020.dtypes
```

```python
# Check the 2021 sales data types.
df_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_df = pd.concat([df_2020, df_2021]).reset_index(drop=True)

combined_df
```

```python
# Check if any values are null.
combined_df.isnull().any()
```

```python
# Check the data type of each column
combined_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
grouped_df = combined_df.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()

grouped_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
grouped_df.head(5)      
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_table = pd.pivot_table(combined_df, index=['region', 'state', 'city'], values=['units_sold'], aggfunc='sum')


# Rename the "units_sold" column to "Total_Products_Sold"
pivot_table.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
pivot_table.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_df_2 = combined_df.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()
grouped_df_2.rename(columns={'total_sales':'Total Sales'}, inplace=True)
# Show the top 5 results.
grouped_df_2.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_table_2 = pd.pivot_table(combined_df, index=['region', 'state', 'city'], values=['total_sales'], aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table_2.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
pivot_table_2.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
filtered_df = combined_df.loc[combined_df['product'] == "Women's Athletic Footwear"]
filtered_df.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
grouped_df_4 = filtered_df.groupby(['retailer','region', 'state', 'city'])['units_sold'].sum().reset_index()
grouped_df_4.rename(columns={'units_sold':'Womens_Footwear_Units_Sold'}, inplace=True)
# Show the top 5 results.
grouped_df_4.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_table_4 = pd.pivot_table(combined_df, index=['retailer','region', 'state', 'city'], values=['units_sold'], aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivot_table_4.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)
# Show the top 5 results.
pivot_table_4.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pivot_table_5 = pd.pivot_table(combined_df, index=['invoice_date'], values=['total_sales'])

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table_5.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the table.
pivot_table_5
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = pivot_table_5.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
daily_sales = daily_sales.sort_values(by='Total Sales', ascending=True)
daily_sales.tail()
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = pivot_table_5.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
weekly_sales = daily_sales.sort_values(by='Total Sales', ascending=True)
weekly_sales.head()
```
